```python
import requests
```

```python
import json

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
```

```python
responseA = requests.get("http://api.nbp.pl/api/exchangerates/tables/A/")
responseB = requests.get("http://api.nbp.pl/api/exchangerates/tables/B/")
responseC = requests.get("http://api.nbp.pl/api/exchangerates/tables/C/")
print("Status code Table A", responseA.status_code)
print("Status code Table B",responseB.status_code)
print("Status code Table C",responseC.status_code)
```

### Description of response parameters for exchange rate queries

1. Table – table type
2. No – table number
3. TradingDate – trading date (for table C only)
4. EffectiveDate – publication date
5. Rates – a list of exchange rates of particular currencies in the table
6. Country – country name
7. Symbol – currency symbol (numerical, concerns historic exchange rates)
8. Currency – currency name
9. Code – currency code
10. Bid – calculated currency buy exchange rate (concerns table C) (for table C only)
11. Ask – calculated currency sell exchange rate (concerns table C) ()
12. Mid – calculated currency average exchange rate (for tables A and B)

```python
jprint(responseA.json())
```

```python
columns = ["Date"]
```

```python
for row in range(len(responseA.json()[0]["rates"])):
    code = responseA.json()[0]["rates"][row]["code"]
    columns.append(code)

print(columns)
```

```python
df = pd.DataFrame(columns=columns)
df
```

```python
df
```

```python
from datetime import datetime, timedelta

# Starting date
start_date = datetime(2020, 1, 1)

# Ending date
end_date = datetime(2023, 12, 23)

# List to store the dates
date_list = []

# Increment by 93 days until the end date
current_date = start_date
while current_date <= end_date:
    date_list.append(current_date.strftime("%Y-%m-%d"))
    current_date += timedelta(days=93)

# Print the list of dates
print(date_list)
```

```python
for start in range(len(date_list)-1):
    response = requests.get("http://api.nbp.pl/api/exchangerates/tables/A/{}/{}".format(date_list[start],date_list[start+1]))
    print(response)
    for iter in range(len(response.json())):
        rows = []
        date = response.json()[iter]["effectiveDate"]
        rows.append(date)
        for row in range(len(response.json()[iter]["rates"])):
            if response.json()[0]["rates"][row]["code"] in columns:
                code = response.json()[0]["rates"][row]["mid"]
                rows.append(code)
        if len(rows) == len(columns)
            df.loc[len(df)] = rows
```

```python
df.tail(10)
```

```python
df.to_csv('exchange_rates.csv', index=False)
```
